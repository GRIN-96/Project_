# 소비자 물가지수 기본금리

```python
df11 = pd.read_excel('/content/품목성질별_소비자물가지수_2020100__20220727184250.xlsx')    # 통계청자료 excel 불러오기
df11
```

```python
df2 = df11.copy() # 원본 훼손을 막기위한 카피
```

```python
df2.pop('시도별')   # 컬럼명 '시도별'을 제외하기
```

```python
df2.rename(columns = {'시점' : '연/월'}, inplace = True)    # 컬럼명 변경 ( 보기 편하게 )

df2.head()
```

```python
df2['연/월'] = df2['연/월'].astype(str)    # 문자열로 변경해 준다.
df2.info()
```

```python
df2= df2.replace('2019.1','2019.10')   #정수 형 형태에서 .10에서 떨어진 0 문자열 상태에서 다시 붙여주기
df2
```

```python
df3 = pd.melt(df2, id_vars=['연/월'], var_name='품목성질', value_name = '물가지수')    # melt 함수로 컬럼구별
df3.head()
```

```python
df3 = df3[df3.품목성질== '총지수']    # 품목 성질이 총지수 인것만 출력
```

```python
df3 = df3.loc[[0,12,20,23,28,30,45,48,50,53,54]]     # 금리 데이터와 비교해야 하기 때문에 자료에서 금리데이터와 같은 연/월 데이터가 오도록 맞춰준다.
df3
```

#기준 금리

```python
!pip install beautifulsoup4
import pandas as pd
import requests
import bs4
import plotly.express as px 
# 기준금리 (2017년 11월~ 2022년 5월)
## 크롤링
url = 'https://www.bok.or.kr/portal/singl/baseRate/list.do?dataSeCd=01&menuNo=200643'

response = requests.get(url)  #요청

html = response.text     

base_rate = bs4.BeautifulSoup(html)

elements = base_rate.find_all('td')    # 답장


## 데이터 프레임 만들기

list1 = [elements[i].text for i in range(0,len(elements))] 

list2 = [' '.join(list1[j:j+3]) for j in range(0,len(list1),3)]

list2 = list2[::-1][39:-1]   # 2017년 데이터부터 필요하기 때문에 인덱싱처리로 자료 정리

date = [a.split(' ')[0] + "-" + a.split(' ')[1] for a in list2]

base_rate = [float(a.split(' ')[3]) for a in list2]


df = pd.DataFrame({'Date' : date, 'BaseRate' : base_rate}) # 데이터프레임
```

```python
df
```

```python
Date_range = {'Date' : {'2017-11월':'2017.11',         # 소비자 물가지수와 같은 x축 표기명을 위해 value명 정정
                        '2018-11월':'2018.11',
                        '2019-07월':'2019.07',
                        '2019-10월':'2019.10',
                        '2020-03월':'2020.03',
                        '2020-05월':'2020.05',
                        '2021-08월':'2021.08',
                        '2021-11월':'2021.11',
                        '2022-01월':'2022.01',
                        '2022-04월':'2022.04',
                        '2022-05월':'2022.05'}}
df1 = df.replace(Date_range)
df1
```
