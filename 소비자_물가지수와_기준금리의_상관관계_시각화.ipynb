#기준 금리 와 기준금리 변동률 구하기

기준금리 자료 출처 : https://www.bok.or.kr/portal/singl/baseRate/list.do?dataSeCd=01&menuNo=200643

```python
!pip install beautifulsoup4
import pandas as pd
import requests
import bs4
import plotly.express as px 
# 기준금리 (2017년 11월~ 2022년 5월)
## 크롤링
url = 'https://www.bok.or.kr/portal/singl/baseRate/list.do?dataSeCd=01&menuNo=200643'

response = requests.get(url)

html = response.text

base_rate = bs4.BeautifulSoup(html)

elements = base_rate.find_all('td')


## 기준금리 데이터 프레임 만들기

list1 = [elements[i].text for i in range(0,len(elements))]

list2 = [' '.join(list1[j:j+3]) for j in range(0,len(list1),3)]

list2 = list2[::-1][39:-1]

date = [a.split(' ')[0] + "-" + a.split(' ')[1] for a in list2]

base_rate = [float(a.split(' ')[3]) for a in list2]


df = pd.DataFrame({'연/월' : date, '기준금리' : base_rate})


## 기준금리 변동률 데이터 프레임 만들기

list_delta = [round(((base_rate[k+1] - base_rate[k])/base_rate[k])*100,1) for k in range(len(base_rate)-1)]  #변동률 구하는 메소드 추가

df4 = pd.DataFrame(list_delta)
```

```python
df
```

```python
Date_range = {'연/월' : {'2017-11월':'2017.11',
                        '2018-11월':'2018.11',
                        '2019-07월':'2019.07',
                        '2019-10월':'2019.10',
                        '2020-03월':'2020.03',
                        '2020-05월':'2020.05',
                        '2021-08월':'2021.08',
                        '2021-11월':'2021.11',
                        '2022-01월':'2022.01',
                        '2022-04월':'2022.04',
                        '2022-05월':'2022.05'}}    ## 가독성을 위해 value값을 소비자 물가지수와 일치하게 변경
df1 = df.replace(Date_range)


df4['Period'] = df1['연/월']    #변동률 차트에도 동일화
df4.rename(columns = {0:'BaseRateChange'}, inplace=True)  # 컬럼명 0 rename
```

# 3. 물가지수(총지수|상품지수|서비스지수) 시각화

```python
# x축, y축 항목 리스트로 정리 

month = df_['연/월']
p_index = df_['총지수']
service = df_['서비스']
product = df_['상품']

# 그래프 
graph3 = go.Figure()

graph3.add_trace(go.Scatter(x = month, y = p_index, name='총지수',
                            mode='lines+markers',
                            line = dict(color = 'grey', width=4)))
graph3.add_trace(go.Scatter(x = month, y = service, name='서비스',
                            line = dict(color = 'blue', width=2))) # dash options include 'dash', 'dot', and 'dashdot'
graph3.add_trace(go.Scatter(x = month, y = product, name='상품',
                            line = dict(color = 'green', width=2)))

graph3.update_layout(title_text = '<b>소비자 물가지수 시각화(총지수|서비스|상품)</b>',
                     xaxis_title = '<b>연/월</b>',
                     yaxis_title = '<b>소비자 물가지수</b>')
graph3.show()                               
```

# 4. 물가지수(서비스지수|하위지수) 시각화
